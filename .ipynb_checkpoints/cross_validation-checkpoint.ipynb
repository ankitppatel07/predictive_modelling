```python
from sklearn import datasets, tree
data = datasets.load_iris()
features = data.data
feature_names = data.feature_names
target = data.target
target_names = data.target_names
labels = target_names[target]
```

```python
tr = tree.DecisionTreeClassifier(min_samples_leaf=10)
tr.fit(features, labels)
```

```python
import graphviz, numpy as np
tree.export_graphviz(tr, feature_names = feature_names, rounded = True, out_file = 'decision.dot')
graphviz.Source(open('decision.dot').read())
```

```python
prediction = tr.predict(features)
print(type(prediction))
print("Accuracy: {:.1%}".format(np.mean(prediction == labels)))
```

```python
#Splitting the data into half and taking the first half of the data
#as training data and the second half as testing data
features_train = features[:int(len(features)/2)]
features_test = features[int(len(features)/2):]
#tr1 = tree.DecisionTreeClassifier(min_samples_leaf=10)
tr.fit(features_train, labels[:int(len(labels)/2)])
tree.export_graphviz(tr, feature_names = feature_names, rounded = True, out_file = 'decision1.dot')
graphviz.Source(open('decision1.dot').read())
prediction1 = tr.predict(features_test)
print("Accuracy when splitting dataset and \ntaking 50-50 training and test data: {:.1%}".format(np.mean(prediction1 == labels[int(len(labels)/2):])))
```

```python
#leave one out cross(LOOC) validation
prediction = []
for i in range(len(features)):
    train_features_looc = np.delete(features, i, axis = 0)
    train_labels_looc = np.delete(labels, i, axis = 0)
    tr.fit(train_features_looc, train_labels_looc)
    prediction.append(tr.predict([features[i]]))
prediction = np.array(prediction)
#print(prediction) #use ravel to flatten the array
print("The Accuracy is: {:.1%}".format(np.mean(prediction.ravel()==labels)))

# LOOC using scikit learn
from sklearn import model_selection

prediction_sklearn = model_selection.cross_val_predict(tr, features, labels, cv=model_selection.LeaveOneOut())
print("The Accuracy is (using sklearn): {:.1%}".format(np.mean(prediction_sklearn ==labels)))
```

```python
#k-folds cross validation
from sklearn import model_selection

#took 5 folds, therefore, data is split into 5, one out of all 5 splits is considered as test data at a time
#so 80% = training data and 20% data = test data and 5 models are learned
k_folds_prediction = model_selection.cross_val_predict(tr, features, labels, cv = model_selection.KFold(n_splits = 5))
print("The Accuracy when using k-folds Cross Validation is (using sklearn): {:.1%}".format(np.mean(prediction_sklearn ==labels)))
```
